Before we handle missing values, let's see statistical summary of this dataset

```python
#Columns had missing values
column_had_miss = df_1.loc[:,df_1.isnull().any()]
column_had_miss.describe()
```

As we observed, the mean and median values of the columns are almost equal, so I decided to fill in the null values with the median value. The reason I didn't use the mean value is to ensure the meaning of the data and avoid confusion (for example: CouponUsed is 1.75 or OrderCount is 3.008)


```python
#Choose Numeric Columns
numeric_cols = df_1.select_dtypes(include='number').columns
#Fill with median value
df_1[numeric_cols] = df_1[numeric_cols].fillna(df_1[numeric_cols].median())
```

```python
#Check Null Again
df_1.isna().sum()
```

**Edit the data in column 'PreferredPaymentMode'**

```python
#Check values in PreferredPaymentMode column
df_1['PreferredPaymentMode'].unique()
```

This column has values that have the same meaning, so I edited them('Cash on Delivery' = 'COD', 'Credit Card' = 'CC').

```python
#ReplaceValues
df_1['PreferredPaymentMode'] = df_1['PreferredPaymentMode'].replace({'Cash on Delivery':'COD','Credit Card':'CC'})
```
